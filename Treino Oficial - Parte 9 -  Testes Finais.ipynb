```python
import gensim
import codecs
import numpy as np
import matplotlib.pyplot as plt
import time
import editdistance
from sklearn.metrics.pairwise import cosine_similarity

path = '/home/lili/Desktop/Projeto/treino/'

modelen = gensim.models.Word2Vec.load(path + 'model-english')
modelpt = gensim.models.Word2Vec.load(path + 'model-portuguese')

h = codecs.open(path + 'on_model_en.txt', encoding = 'utf-8', mode = 'r')
i = codecs.open(path + 'on_model_pt.txt', encoding = 'utf-8', mode = 'r')

words_en = h.read().split()
words_pt = i.read().split()

alpha1 = 10**(-16)
alpha2 = 10**(-11)
alpha3 = 10**(-11)
alpha4 = 10**(-13)
alpha5 = 10**(-10)
```

```python

W1 = np.random.rand(100,100)
W3 = np.random.rand(100,100)

for k in range(3311):
    X = modelpt[words_pt[k]][np.newaxis]
    y = modelen[words_en[k]][np.newaxis]
    #print X
    #X é o vetor input, yo é o calculado e y é o output esperado
    y1 = np.dot(X,W1)
    z = np.dot(X,W3)
    y3 = 1/(1+np.exp(-z)) #função sigmoide
    sigd = np.exp(-z)/((1+np.exp(-z))**2) #derivada da função sigmoide
    
    dif1 = y - y3
    dif3 = np.multiply((y - y3),sigd)
    
    djdw1 = np.dot(np.transpose(X),dif1)
    djdw3 = np.dot(np.transpose(X),dif3)

    W1 = W1 - djdw1*alpha1
    W3 = W3 - djdw3*alpha3
```

```python
#Teste distancia: edit distance
f1 = codecs.open(path + 'dist6.txt', encoding = 'utf-8', mode = 'w')

edit = []

for i in range(1500):
    X = modelpt[words_pt[i + 3311]]
    z1 = np.dot(X,W6)
    word = modelen.most_similar([z1])[0][0]
    v = editdistance.eval(word, words_en[i + 3311])
    edit.append(v)
edit.sort()

for i in range(1500):
    f1.write(str(edit[i]))
    f1.write('\n')
f1.close()        
```

```python
#Teste distancia: similarity
f1 = codecs.open(path + 'distsim6.txt', encoding = 'utf-8', mode = 'w')

edit = []
for i in range(1500):
    X = modelpt[words_pt[i + 3311]]
    z1 = np.dot(X,W6)
    word = modelen.most_similar([z1])[0][0]
    v = modelen.similarity(word, words_en[i + 3311])
    edit.append(v)

edit.sort()

for i in range(1500):
    f1.write(str(edit[i]))
    f1.write('\n')
f1.close()        
```
